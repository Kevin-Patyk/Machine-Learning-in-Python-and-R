# K-Nearest Neighbors (K-NN)

## Importing the libraries

```python
#numpy
import numpy as np
#matplotlib
import matplotlib.pyplot as plt
#pandas
import pandas as pd
```

## Importing the dataset

```python
#import the data
dataset = pd.read_csv('Social_Network_Ads.csv')
#selecting the predictor matrix
X = dataset.iloc[:, :-1].values
#selecting the target variable
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
#using sklearn for the train/test split
from sklearn.model_selection import train_test_split
#performing the split with random state 0 and 25% in test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#displaying the predictor matrix
print(X_train)
```

```python
#displaying the target variable
print(y_train)
```

```python
#displaying the test predictor matrix
print(X_test)
```

```python
#displaying the test outcome variable
print(y_test)
```

## Feature Scaling

```python
#using sklearn for scaling
from sklearn.preprocessing import StandardScaler
#making an object of class standard scaler
sc = StandardScaler()
#scaling the training predictors
X_train = sc.fit_transform(X_train)
#scaling the testing predictors 
X_test = sc.transform(X_test)
```

```python
#displaying after scaling
print(X_train)
```

```python
#displaying after scaling
print(X_test)
```

## Training the K-NN model on the Training set

```python
#using the module KNN from sklearn
from sklearn.neighbors import KNeighborsClassifier
#making an object of class KNN with 5 neighbors, Minkowski distance, and power metric 2 (which is for euclidean)
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
#fitting the model
classifier.fit(X_train, y_train)
```

## Predicting a new result

```python
#printing a new results
print(classifier.predict(sc.transform([[30,87000]])))
```

## Predicting the Test set results

```python
#making a new prediction based off of the testing set
y_pred = classifier.predict(X_test)
#comparing the predictions of the predicted outcome and the actual outcome
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```python
#using sklearn modules for the confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
#making the confusion matrix 
cm = confusion_matrix(y_test, y_pred)
#displaying the confusion matrix
print(cm)
#calculating the accuracy
accuracy_score(y_test, y_pred)
```

## Visualising the Training set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
